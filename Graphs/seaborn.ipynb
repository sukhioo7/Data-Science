```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv('https://github.com/sukhioo7/Data-Science/blob/main/Graphs/tips.csv?raw=ture')
```

```python
tips.head()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',palette='mako')
plt.title('Total Bill VS Tips',size=30)
plt.xlabel('Total Bill',size=16)
plt.ylabel('Tips',size=16)
plt.grid(alpha=0.2)
plt.show()
```

```python
tips['Gender'].value_counts()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(data=tips,x='Gender',hue='time')
```

```python
plt.figure(figsize=(12,7))
sns.barplot(data=tips,x='Gender',y='total_bill')
```

```python
plt.figure(figsize=(12,7))
sns.histplot(data=tips,x='total_bill',bins=40,kde=True)
```

```python
plt.figure(figsize=(12,7))
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(data=tips,x='day',y='tip')
```

```python
plt.figure(figsize=(12,7))
sns.violinplot(data=tips,x='day',y='tip',hue='Gender',split=True)
```

```python
sns.pairplot(tips,hue='Gender')
```
