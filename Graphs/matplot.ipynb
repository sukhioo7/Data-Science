```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temp_avg = [9,12,15,22,27,35,38,30,27,24,18,10]
temp_min = [7,10,12,20,24,33,35,28,25,22,16,8]
temp_max = [11,14,18,25,30,38,40,33,29,26,20,14]
months = 'Jan Feb Mar Apr May Jun Jul Aug Sep Oct Nov Dec'.split()
```

```python
plt.figure(figsize=(13,7))
plt.plot(months,temp_avg,color='#0CBABA',linewidth=1.2,alpha=0.8,marker='o',markersize=3,label='AVG')
plt.plot(months,temp_min,color='#A40606',linewidth=1.2,marker='o',linestyle='-.',markersize=4,label='MIN')
plt.plot(months,temp_max,color='#FF0000',linewidth=1.2,marker='o',markersize=4,label='MAX')
plt.xlabel('Months',color='#645394',size=12)
plt.ylabel('Temp',color='#645394',size=12)
plt.grid(alpha=0.4)
plt.legend(shadow=0.3)
plt.title('Months VS Temp',size=30)
plt.show()
```

```python
w = 0.4

avg_points = np.arange(len(months))

min_points = avg_points + w
```

```python
plt.figure(figsize=(15,7))
# plt.plot(months,temp_avg,color='#0CfABA',linewidth=1.2,alpha=0.8,marker='o',markersize=3,label='AVG')
plt.bar(avg_points,temp_avg,color='#0CBABA',width=w)
plt.bar(min_points,temp_min,color='#A40606',width=w)

plt.grid(alpha=0.2)
plt.title('Temp VS Months',size=45)
plt.xlabel('Months',size=15)
plt.ylabel('Temp',size=15)
plt.xticks(min_points-0.2,months)
plt.show()
```
