# Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the dataset

```python
country = pd.read_csv('https://github.com/sukhioo7/dataset/blob/main/002%203.01.Country-clusters.csv?raw=true')
```

**Displaying the first few rows of the dataset**

```python
country.head()
```

# Visualizing the geographical distribution of countries

```python
plt.figure(figsize=(13, 7))
plt.scatter(country['Longitude'], country['Latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Latitude and Longitude of Countries')
plt.show()
```

# Performing K-means clustering on geographical coordinates

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(country[['Latitude','Longitude']])
```

# Adding cluster labels to the dataset

```python
country['Clusters'] = kmeans.predict(country[['Latitude','Longitude']])
```

```python
country
```

# Visualizing the clustering results

```python
plt.figure(figsize=(13, 7))
plt.scatter(country['Longitude'], country['Latitude'],c=country['Clusters'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Latitude and Longitude of Countries')
plt.show()
```

# Determining the optimal number of clusters using the elbow method

```python
wcss = []
for k in range(2,6):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(country[['Longitude','Latitude']])
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

# Plotting the within-cluster sum of squares (WCSS) for different values of K

```python
plt.figure(figsize=(13, 7))
plt.plot(range(2,6),wcss)
plt.xlabel('Value of K')
plt.ylabel('WCSS')
plt.title('Best Value for the Cluster')
plt.show()
```

# Loading customer dataset

```python
customers = pd.read_csv('https://github.com/sukhioo7/dataset/blob/main/011%203.12.Example.csv?raw=True')
```

# Displaying the first few rows of the customer dataset

```python
customers.head()
```

# Visualizing the satisfaction and loyalty of customers

```python
plt.figure(figsize=(13, 7))
plt.scatter(customers['Satisfaction'], customers['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.title('Scatter Plot of Satisfaction and Loyalty of Customers')
plt.show()
```

# erforming K-means clustering on customer satisfaction and loyalty

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(customers)
```

```python
customers['Clusters'] = kmeans.predict(customers)
```

```python
customers.head()
```

# Visualizing the clustering results for customer satisfaction and loyalty

```python
plt.figure(figsize=(13, 7))
plt.scatter(customers['Satisfaction'], customers['Loyalty'],c=customers['Clusters'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.title('Scatter Plot of Satisfaction and Loyalty of Customers')
plt.show()
```

# Scaling the features and performing K-means clustering

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
```

```python
standard_scaler.fit(customers[['Satisfaction','Loyalty']])
```

```python
scaled_customers = standard_scaler.transform(customers[['Satisfaction','Loyalty']])
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(scaled_customers)
```

```python
customers['Clusters'] = kmeans.predict(scaled_customers)
```

```python
customers.head()
```

# Visualizing the clustering results for scaled customer satisfaction and loyalty

```python
plt.figure(figsize=(13, 7))
plt.scatter(customers['Satisfaction'], customers['Loyalty'],c=customers['Clusters'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.title('Scatter Plot of Satisfaction and Loyalty of Customers')
plt.show()
```

# Determining the optimal number of clusters using the elbow method

```python
error = []

for k in range(2,10):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(customers[['Satisfaction','Loyalty']])
  error.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(13, 7))
plt.plot(range(2,10),error)
plt.xlabel('Value of K')
plt.ylabel('WCSS')
plt.title('Best Value for the Cluster')
plt.show()
```
